```python
import tensorflow as tf
import cv2 
import imghdr
import os
```

```python
def getData(wType):
    allowed = ["train", "test"]
    if wType not in allowed:
        return False
    return {
        "direc": f"data/{wType}/",
        "annotation": f"data/{wType}.csv"
    }
```

```python
image_allowed = ['jpg', 'jpeg', 'bmp', 'png']
for image_class in os.listdir(getData('train')["direc"]):
    for image in os.listdir(os.path.join(getData('train')["direc"], image_class)):
        img_path = os.path.join(getData('train')["direc"], image_class, image)
        try:
            img = cv2.imread(img_path)
            tip = imghdr.what(img_path)
            if tip not in image_allowed:
                print("Image not allowed for this format")
                os.remove(img_path)
        except Exception as e:
            print("Error:".format(img_path))
            
    
```
